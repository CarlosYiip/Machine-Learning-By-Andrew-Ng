```python
# Part1: Initialization
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model

x, y = np.loadtxt(fname='ex1data1.txt', delimiter=',', unpack=True, ndmin=1)
m, n = len(x), 1
data_x = np.zeros((m, n))
data_y = np.zeros((m , 1))
for i in range(len(x)):
    data_x[i] = x[i]
    data_y[i] = y[i]
regr = linear_model.LinearRegression()
regr.fit(data_x, data_y);
```

```python
# Part2: Plotting the sample data
plt.scatter(data_x, data_y)
plt.show()
```

```python
# Part3: Gradient descent
def computeCost(X, Y, theta):
    J = (X @ theta - Y).T @ (X @ theta - Y) / (2 * len(Y))
    return J

def gradientDescent(X, Y, theta, alpha, num_iters):
    m = len(y)
    J_history = np.zeros((num_iters, 1))
    for i in range(num_iters):
        theta -= alpha / m * ((X @ theta - Y).T @ X).T
        J_history[i] = computeCost(X, Y, theta)
    return theta, J_history

X = np.zeros((m, n + 1))
theta = np.zeros((2, 1))
iterations = 1500;
alpha = 0.01
for i in range(len(data_x)):
    X[i] = 1, data_x[i]

theta, J_history = gradientDescent(X, data_y, theta, alpha, iterations)
predict_y1 = X @ theta
predict_y2 = regr.predict(data_x)

predict1 = np.array([1, 3.5]) @ theta * 10000
predict2 = np.array([1, 7]) @ theta * 10000
print('Theta found by gradient descent: {}, {}'.format(theta[0][0], theta[1][0]))
print('For population = 35,000, we predict a profit of {}'.format(predict1[0]))
print('For population = 70,000, we predict a profit of {}'.format(predict2[0]))

plt.scatter(data_x,data_y)
plt.plot(data_x, predict_y1, color="red")
plt.plot(data_x, predict_y2, color="yellow")
plt.show()
```

# EX1_multi

```python
# Part1: Data preprocessing and feature normalization
import numpy as np
from sklearn import linear_model

def featureNormalise(X):
    mu = np.zeros((len(X[0]), 1))
    sigma = np.zeros((len(X[0]), 1))
    for i in range(len(X[0])):
        mu[i] = np.mean(X[:, i:i+1])
        sigma[i] = np.std(X[:, i:i+1])
        X[:, i:i+1] = (X[:, i:i+1] - mu[i]) / sigma[i]
    return mu, sigma
            
raw_data = np.loadtxt('ex1data2.txt', delimiter=',')
X, Y = raw_data[:, :2].copy(), raw_data[:, -1:].copy()

mu, sigma = featureNormalise(X)
X = np.insert(X, 0, 1, axis=1)
```

```python
# Part2: Multivariable Gradient Descent
def computeCostMulti(X, Y, theta):
    return (X @ theta - Y).T @ (X @ theta - Y) / (2 * m)

def gradientDescentMulti(X, Y, theta, alpha, num_iters):
    m = len(Y)
    J_history = np.zeros((num_iters, 1))
    for i in range(num_iters):
        theta -= alpha / m * ((X @ theta - Y).T @ X).T
        J_history[i] = computeCostMulti(X, Y, theta)
    return theta, J_history

theta = np.zeros((3, 1))
alpha = 1.25
num_iters = 1500

theta, J_history = gradientDescentMulti(X, Y, theta, alpha, num_iters)
test_x = np.zeros((1, 3))
test_x[0][0], test_x[0][1:] = 1, (([1650, 3] - mu.T) / sigma.T)[0]
predict_price = test_x @ theta
print('Predict price for Predicted price of a 1650 sq-ft, 3 br house: {}\n'.format(int(predict_price)))

print('Cost vs Iteration')
iteration = np.linspace(1, 1500, 1500)
plt.plot(iteration, J_history)
plt.show()
```

```python
#Part3: Using sklearn module
regr = linear_model.LinearRegression(normalize=True)
regr.fit(raw_data[:, :2], raw_data[:, -1:])
predict_price = regr.predict(np.array([1650, 3]).reshape(1, -1))
print('Predicted price for same house by using sklearn module: {}'.format(predict_price[0][0]))
```

```python
# Part4 Normal Equations
X_NE, Y_NE = raw_data[:, :2].copy(), raw_data[:, -1:]
X_NE = np.insert(X_NE, 0, 1, axis=1)
thetaNE = np.linalg.pinv(X_NE.T @ X_NE) @ X_NE.T @ Y_NE
print('Predicted price for same house by using normal equation: {}'.format((np.array([1, 1650, 3]) @ thetaNE)[0]))
```
